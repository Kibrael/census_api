```python
import pandas as pd
```

```python
msa_data = pd.read_csv("data/list1.csv", dtype="object", encoding='latin-1')
msa_data
```

```python
msa_data.columns
```

```python
cols = ["CBSA Code", "CBSA Title"]
msa_df = msa_data[cols][msa_data["Metropolitan/Micropolitan Statistical Area"]=="Metropolitan Statistical Area"]
msa_df.head()
```

```python
msa_df = msa_df[msa_df.duplicated(["CBSA Code", "CBSA Title"])==False]
print(len(msa_df))
msa_df
```

```python
import csv

csv_filename = "2017HMDAMSADescription.txt"
 
with open("output/"+csv_filename, 'w') as f:
    #writer = csv.writer(f ,delimiter='\t')
    #writer.writerows(msa_df["CBSA Code", "CBSA Title"])
    for index, row in msa_df.iterrows():
        line = row["CBSA Code"]+"\t"+row["CBSA Title"]
        line = line + "\n"
        print(line)
        f.write(line)
```
